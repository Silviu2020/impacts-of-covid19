## 2. Create a function for the cleaning process

```python
def cleaning(df_name, name_in_col, path_out):
    '''This function transforms the csv files that we downloaded from https://finance.yahoo.com/'''
    # set the column Date to datetime
    df_name['Date'] = pd.to_datetime(df_name['Date'])
    # create a new col
    df_name = df_name.assign(Name=name_in_col)
    # save the DataFrame
    df_name.to_csv(path_out, index=False)
    # test it
    test = pd.read_csv(path_out)
    return test.head()
```

## 3. Clean the csvs one-by-one

```python
real_estate_df = pd.read_csv('Data/FREL (Real Estate).csv')
cleaning(real_estate_df, 'Real Estate', 'Data/real_estate_df.csv')
```

```python
zoom_df = pd.read_csv('Data/ZM (Zoom).csv')
cleaning(zoom_df, 'Zoom', 'Data/zoom_df.csv')
```

```python
netflix_df = pd.read_csv('Data/NFLX (Netflix Stock).csv')
cleaning(netflix_df, 'Netflix', 'Data/netflix_df.csv')
```

```python
german_market_index_df = pd.read_csv('Data/^GDAXI (German Market Index).csv')
cleaning(german_market_index_df, 'DAX', 'Data/german_market_index_df.csv')
```

```python
sp_500_df = pd.read_csv('Data/^GSPC (S&P 500).csv')
cleaning(sp_500_df, 'S&P 500', 'Data/sp_500_df.csv')
```

```python
vix_df = pd.read_csv('Data/^VIX (Volatility Index).csv')
cleaning(vix_df, 'VIX', 'Data/vix_df.csv')
```
